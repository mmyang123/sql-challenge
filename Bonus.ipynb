```python
import pandas as pd
# SQLALchemy
from sqlalchemy import create_engine
from sql_pass import postgresql_pwd
import matplotlib.pyplot as plt
import numpy as np
# Create an engine that can talk can talk to the database
engine = create_engine('postgresql://postgres:' + postgresql_pwd + '@localhost:5432/sql_challenge_db')
connection = engine.connect()
```

```python
salary_df = pd.read_sql('select * from employees e inner join salaries s on e.emp_no = s.emp_no inner join titles t on e.emp_title = t.title_id', connection)
```

```python
salary_df.plot.hist(column='salary', bins=100, figsize=(10,8), grid=True, orientation='horizontal')
plt.ylabel('salary')
```

```python
avg_salaries_df = salary_df.groupby('title')['salary'].mean()
avg_salaries_df.plot.bar(x='title',y='salary',rot=45)
plt.ylabel('Salary')
```
